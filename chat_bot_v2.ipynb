```python
import time
import json
import re
from difflib import get_close_matches

def bot_says(text):
    """Simulate typing effect for chatbot output."""
    for char in text:
        print(char, end='', flush=True)
        time.sleep(0.02)
    print()

print("\n Courssistant Bot (type 'bye' to exit)\n")

# Load memory (if exists)
try:
    with open("course_bot_memory.json", "r") as file:
        memory = json.load(file)
        memory_questions = memory["questions"]
        memory_responses = memory["responses"]
except:
    memory_questions = []
    memory_responses = []

def get_best_match(user_input):
    """Find closest matching question from memory."""
    all_known = memory_questions
    match = get_close_matches(user_input, all_known, n=1, cutoff=0.6)
    return match[0] if match else None

# --- MAIN CHAT LOOP ---
while True:
    user_input = input("You: ").lower().strip()

    if user_input == "bye":
        bot_says("Saving memory... âœ…")
        with open("course_bot_memory.json", "w") as file:
            json.dump({"questions": memory_questions,
                       "responses": memory_responses}, file)
        bot_says("Goodbye! ðŸ‘‹ BURN THE MIDNIGHT OIL BUANA AND LOCK IN !!!")
        break

    # âœ… Auto-correct / best guess
    if user_input not in memory_questions:
        guess = get_best_match(user_input)
        if guess:
            bot_says(f"Did you mean: '{guess}'? (y/n)")
            confirm = input("You: ").lower()
            if confirm == "y":
                user_input = guess

    # âœ… Respond from memory
    if user_input in memory_questions:
        reply = memory_responses[memory_questions.index(user_input)]
        bot_says(reply)
        continue

    # --- 5 INTENTS (AI/COURSE FAQs) ---
    if any(word in user_input for word in ["hello", "hi", "hey"]):
        bot_says("Hello there! ðŸ‘‹ How can I assist you with your AI course today?")

    elif "what is ai" in user_input or "define ai" in user_input:
        bot_says("Artificial Intelligence (AI) is the simulation of human intelligence in machines that can learn, reason, and make decisions.")

    elif "applications of ai" in user_input or "use of ai" in user_input:
        bot_says("AI is used in many fields â€” from healthcare and finance to transportation, education, and robotics. ðŸ¤–")

    elif "who teaches" in user_input or "lecturer" in user_input:
        bot_says("The AI course is taught by qualified lecturers who specialize in machine learning, data science, and programming.")

    elif "topics" in user_input or "syllabus" in user_input:
        bot_says("The AI course covers Machine Learning, Neural Networks, Natural Language Processing, and Robotics fundamentals.")

    elif "contact" in user_input or "support" in user_input:
        bot_says("You can reach the department office via email or visit the academic office during working hours. ðŸ“§")

    elif "exam" in user_input or "assessment" in user_input:
        bot_says("The course assessment includes assignments, CATs, and a final exam. Continuous assessment is key to success!")

    else:
        # âœ… Learn new response
        bot_says("I don't have an answer for that yet ðŸ˜…\nCan you teach me how to respond?")
        new_reply = input("Reply: ")
        memory_questions.append(user_input)
        memory_responses.append(new_reply)
        bot_says("âœ… Got it! Thanks, Iâ€™ve learned something new.")

```
